#Some commonly used Activation functions
##1.1: f(x) = e^x -> f'(x) = e^x
##1.2: f(x) = 0.25*x if x <= 0 else x -> f'(x) = 0.25 if x <= 0 else 1
##1.3: f(x) = 0.25(e^x-1) if x<= 0 else x -> f'(x) = 0.25*e^x if x <= 0 else 1
##1.4: f(x) = log(1+e^x) -> f'(x) = e^x/(1+e^x)
##1.5: f(x) = x/(|x|+1) -> f'(x) = (|x| + 1) - (x/|x|)*x


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_function_and_derivative(x, fx, dfx):
    
    # setting the axes at the centre
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # plot the function
    ax.plot(x, fx, color="#d35400", linewidth=3, label = "function")
    ax.plot(x, dfx, color="#1abd15", linewidth=3, label = "derivative")
    
    ax.legend(loc="upper left", frameon=False)
    plt.show()
```

#Sigmoid

```python
def sigmoid(x):                                        
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    y = sigmoid(x)
    return y*(1-y)

def sigmoid_list(x_data):
    fx = [sigmoid(x) for x in x_data]
    return fx

def d_sigmoid_list(x_data):
    dfx = [sigmoid_derivative(x) for x in x_data]
    return dfx
```

```python
x_data = list(range(-100, 100+1))
x_data = [x/10 for x in x_data]
```

```python
fx = sigmoid_list(x_data)
dfx = d_sigmoid_list(x_data)
plot_function_and_derivative(x_data, fx, dfx)
```

#Exponential

```python
def exponential(x):                                        
    return np.exp(x)

def exponential_derivative(x):
    return exponential(x)

def exponential_list(x_data):
    fx = [exponential(x) for x in x_data]
    return fx

def d_exponential_list(x_data):
    dfx = [exponential_derivative(x) for x in x_data]
    return dfx
```

#PReLU

```python
def prelu(x):                                        
    return 0.25*x if x <= 0 else x

def prelu_derivative(x):
    y = 0.25 if x <= 0 else 1
    return y

def prelu_list(x_data):
    fx = [prelu(x) for x in x_data]
    return fx

def d_prelu_list(x_data):
    dfx = [prelu_derivative(x) for x in x_data]
    return dfx
```

#ELU

```python
def elu(x):                                        
    return 0.25*(np.exp(x)-1) if x <= 0 else x

def elu_derivative(x):
    y = 0.25*np.exp(x) if x <= 0 else 1
    return y

def elu_list(x_data):
    fx = [elu(x) for x in x_data]
    return fx

def d_elu_list(x_data):
    dfx = [elu_derivative(x) for x in x_data]
    return dfx
```

#SOFTPLUS

```python
def softplus(x):                                        
    return np.log(1+np.exp(x))

def softplus_derivative(x):
    y = np.exp(x)/(1+np.exp(x))
    return y

def softplus_list(x_data):
    fx = [softplus(x) for x in x_data]
    return fx

def d_softplus_list(x_data):
    dfx = [softplus_derivative(x) for x in x_data]
    return dfx
```

```python
# EXERCISE 10
x_list = list([-1, 0, 1])
x_gradient_list = d_softplus_list(softplus_list(x_list))
print([round(num, 2) for num in x_gradient_list])
```

#SOFTSIGN

```python
def softsign(x):                                        
    return x/(abs(x)+1)

def softsign_derivative(x):
    y = ((abs(x) + 1) - (x**2/abs(x)))/(abs(x)+1)**2
    return y

def softsign_list(x_data):
    fx = [softsign(x) for x in x_data]
    return fx

def d_softsign_list(x_data):
    dfx = [softsign_derivative(x) for x in x_data]
    return dfx
```

```python
# EXERCISE 11
x_list = list([-1, 0, 1])
x_softsign_list = softsign_list(x_list)
print([round(num, 2) for num in x_softsign_list])
```

```python
# EXERCISE 12
x_list = list([-1, 0.001, 1])
x_gradient_list = d_softsign_list(softsign_list(x_list))
print([round(num, 2) for num in x_gradient_list])
```

# SIMPLE OPTIMIZATION

```python
import random

def f(x):
    return 3*x**4 - 4*x**2 - 6*x - 3

def gradient(f, x, epsilon):
    return (f(x+epsilon/2) - f(x-epsilon/2)) / epsilon

x = 3.0
print("initial x: ",x)

def find_minimum(f, x, num_iteration, step):
  for _ in range(num_iteration):
    dx = gradient(f, x, epsilon = 1e-10)
    if dx > 0:
      x = x - step
    elif dx < 0:
      x = x + step
    else:
      x = x   
  return x 
```

```python
# EXERCISE 13
x = find_minimum(f = f, x = x, num_iteration = 1, step = 0.1)
print(x)
```

```python
# EXERCISE 14
x = find_minimum(f = f, x = 3, num_iteration = 5, step = 0.1)
print(round(x, 2))
```

```python
# EXERCISE 15
x = find_minimum(f = f, x = 3, num_iteration = 100, step = 0.2)
print(round(x, 2))
```
