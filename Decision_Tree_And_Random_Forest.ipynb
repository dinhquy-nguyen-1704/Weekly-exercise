```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
data_path = "./Housing.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.to_list()
print(categorical_cols)
```

```python
ordinal_encoder = OrdinalEncoder()
encoded_categorical_cols = ordinal_encoder.fit_transform(df[categorical_cols])

encoded_categorical_df = pd.DataFrame(encoded_categorical_cols, columns = categorical_cols)
numerical_df = df.drop(categorical_cols, axis =1)
encoded_df = pd.concat([numerical_df, encoded_categorical_df], axis=1)
```

```python
encoded_df.head()
```

```python
normalizer = StandardScaler()
dataset_arr = normalizer.fit_transform(encoded_df)
```

```python
X, y = dataset_arr[:, 1:], dataset_arr[:, 0]
```

```python
test_size=0.3
random_state=1
is_shuffle=True
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=test_size, random_state=random_state, shuffle=is_shuffle)
```

```python
regressor_dt = DecisionTreeRegressor(random_state=random_state)
regressor_dt.fit(X_train, y_train)
```

```python
regressor_rf = RandomForestRegressor(random_state=random_state)
regressor_rf.fit(X_train, y_train)
```

```python
y_dt_pred = regressor_dt.predict(X_val)
y_rf_pred = regressor_rf.predict(X_val)
```

```python
mae_dt = mean_absolute_error(y_val, y_dt_pred)
mae_rf = mean_absolute_error(y_val, y_rf_pred)
mse_dt = mean_squared_error(y_val, y_dt_pred)
mse_rf = mean_squared_error(y_val, y_rf_pred)
```

```python
print('Evaluation results on validation set:')
print(f'Mean Absolute Error of DecisionTreeRegressor: {round(mae_dt, 4)}')
print(f'Mean Absolute Error of RandomForestRegressor: {round(mae_rf, 4)}')
print(f'Mean Squared Error of DecisionTreeRegressor: {round(mse_dt, 4)}')
print(f'Mean Squared Error of RandomForestRegressor: {round(mse_rf, 4)}')
```
